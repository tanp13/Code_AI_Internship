```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Generation', y='Total', palette='coolwarm')
plt.title('Total Stats by by Generation')
plt.xlabel('Generation')
plt.ylabel('Total Stats')
plt.show()
```

```python
legendary_counts = df['Legendary'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(legendary_counts, labels=legendary_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Proportion of Legendary Pokemon')
plt.show()
```

```python
type1_counts = df['Type 1'].value_counts()

#set the color palette to "pastel"
sns.set_palette('pastel')

#create the pie chart
plt.figure(figsize=(10,8))
plt.pie(type1_counts, labels=type1_counts.index, startangle=90, counterclock=False, autopct='%1.1f%%')
plt.title('Distribution of Pokemon Type 1')
plt.axis('equal')
plt.show()
```

```python
generation_totals = df.groupby('Generation')['Total'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=generation_totals, x='Generation', y='Total', marker='o')
plt.title('Mean Total Stats by Generation')
plt.xlabel('Generation')
plt.ylabel('Mean Total Stats')
plt.xticks(generation_totals['Generation'])
plt.grid(True)
plt.show()
```

```python
# Selecting only the numerical columns for correlation
numerical_columns = df[['Total','HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']]

# Calculating the correlation matrix
correlation_matrix = numerical_columns.corr()

#creating a heatmap
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidth=.5)
plt.title('Correlation Matrix of pokemon stats')
plt.show()
```

FEATURE ENGINEERING

```python
#create new features
#total stats without hp
df['Total_without_HP']= df['Total']-df['HP']
```

```python
df.head()
```

```python
#normalising features
from sklearn.preprocessing import MinMaxScaler

#normalize attac,defense,speed
scaler = MinMaxScaler()
df[['Attack_scaled','Defense_scaled','Speed_scaled']] = scaler.fit_transform(df[['Attack','Defense','Speed']])
```

```python
df.head()
```

```python
#bining

bins = [0,50,100,150,200]
labels=['Very Slow','Slow','Fast', 'Very Fast']
df['Speed_Category'] = pd.cut(df['Speed'], bins=bins, labels=labels)
```

```python
df.head()
```

```python
fast_pokemons = df[df['Speed_Category'] == 'Fast']
print(fast_pokemons)
```

```python
#label encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Legendary_Encoded'] = le.fit_transform(df['Legendary'])
```
